```python
# 필요 라이브러리 호출
import glob
import pandas as pd
```

```python
# Windows 11 환경과 달리 MacOS에는 파일명 정렬이 안된 채로 glob가 되기 때문에 sorted 메서드를 활용해야 한다.
meeting_rec_tmp = glob.glob('../First_project/dataes/main_congress_meeting_records/*.csv')
meeting_rec = sorted(meeting_rec_tmp)
```

```python
# 데이터프레임 생성 -> concat -> 결과 확인
df = pd.DataFrame()
for i in meeting_rec:
    fil = pd.read_csv(i)
    df = pd.concat([df, fil])
df
```

```python
# 불필요 컬럼 제거
del df['Unnamed: 0']
df
```

```python
# 파일 저장
df.to_csv('../First_project/dataes/03_21st_main_conf_total.csv')
```
