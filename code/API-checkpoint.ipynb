# openai api 요약

```python
from openai import openai_api_key # OpenAI API 키 import
import openai # OpenAI 라이브러리 import

# OpenAI API 키 설정
openai.api_key = openai_api_key

# 'contents' 열의 내용을 요약하여 새로운 열에 추가하는 함수
def summarize_content(content):
    response = openai.Completion.create(
      engine="davinci",
      prompt=content,
      max_tokens=50, # 원하는 요약 길이 설정
      temperature=0.5, # 요약의 다양성을 조절하는 매개변수
      n = 1 # 요약을 몇 개 생성할지 설정
    )
    summary = response.choices[0].text.strip()
    return summary

# 'contents' 열의 각 내용을 요약하여 새로운 열에 추가
df['summary'] = df['contents'].apply(summarize_content)
```

# 국회의원 이름으로 검색

```python
def search_bill_content(df, proposer_name):

    
    # 결과를 담을 빈 리스트 생성
    result = []
    
    # 데이터프레임 순회
    for index, row in df.iterrows():
        # PROPOSER에 있는 값과 PUBL_PROPOSER에 있는 값이 모두 포함되는지 확인
        if proposer_name in row['PROPOSER'] or any(proposer_name in name for name in row['PUBL_PROPOSER']):
            # 만족하는 경우 법안 내용과 BILL_NAME을 결과 리스트에 추가
            result.append({'법안': row['BILL_NAME'], '법안내용': row['contents']})
    
    return result
```

```python
from fastapi import FastAPI
from typing import Optional
import pymysql

# MySQL 연결 설정
conn = pymysql.connect(
    host='your_host',
    port=3306,
    user='your_username',
    password='your_password',
    database='your_database'
)

# FastAPI 애플리케이션 설정
app = FastAPI()

# getBill 함수 정의
def getBill(bill_id):
    try:
        with conn.cursor() as cursor:
            # SQL 쿼리 실행
            sql = "SELECT * FROM your_table WHERE BILL_ID = %s"
            cursor.execute(sql, (bill_id,))
            result = cursor.fetchone()  # 결과 가져오기
            return result
    except Exception as e:
        return f"에러 발생: {e}"
    finally:
        conn.close()  # 연결 종료

# FastAPI 엔드포인트 설정
@app.get("/bill/{bill_id}")
def read_bill(bill_id: str):
    result = getBill(bill_id)
    if result:
        return result
    else:
        return {"message": "해당하는 법안이 없습니다."}
```
